# Introduction to Data Science for Librarians

By Sarah W. Sutton, Ph.D.

## Preface
This open textbook was written for a course by the same name taught in the School of Library and Informaton Management at Emporia State University. It assumes no prior knowledge of data science, computer science, or programming.

The course and this textbook have their roots in a personal learning project I  completed during the COVID-19 pandemic, studying data analyitics at Kansas State University. I studied programming languages like R and Python, data analytics including statistical analysis of data, how to work with, study, and analyze big data, that is "data sets that are too big to be dealt with by traditional data-processing software" ("Big data", 2025) and computing systems. The term data science incorporates all those things.

My initial data science education was focused on business applications of data science. Most of the problems I worked on were things like analyzing tweets to determine how consumers felt about a particular product. But the more I learned, the more I realized that this was something that librarians should know about. I even brought some of the research problems I was working on at the time to
the work and projects I was doing in my data science courses. My idea that this was something that librarians should know about isn’t an original one. The LIS field has taken hold of the idea, which turns out to have been very useful in terms of my planning and preparation for this course.

Data science is also a VERY big topic and this book is an introduction. It’s meant not to teach you everything there is to know about data science, or even just everything there is to know about the intersection of data science and library and information. Rather, I intend to give you a basis from which to keep learning about data science.

This book approaches learning about data science not only from the perspective of its application by libraries but also libraries’ developing programs and services to support
researchers in all disciplines to capture, store, wrangle, analyze, preserve, and make available to other researchers their data.
